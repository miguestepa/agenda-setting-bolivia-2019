```python
emotion_dict = pd.read_csv('lexicon_spanish_fixed.csv', encoding='latin-1', sep = ';')
```

```python
df_emotion = pd.DataFrame(emotion_dict)
```

```python
df_emotion
```

```python
df_emotion = df_emotion.drop(['Column1', 'word','emotion-intensity-score'], axis=1)
```

```python
df_emotion
```

```python
for word in range(len(df_emotion.index)):
    #print(word)
    #print(df_emotion.iloc[word, 0])
    if df_emotion.iloc[word, 0][-1] == 'o':
        #print(df_emotion.iloc[word, 0])
        before_change = df_emotion.iloc[word, 0]       
        changing = list(before_change)
        changing[-1] = 'a'
        after_change = "".join(changing)
        #print(after_change)
        new_row = {'Spanish-es':after_change, 'emotion':df_emotion.iloc[word, 1]}
        df_emotion = df_emotion.append(new_row, ignore_index=True)
    elif df_emotion.iloc[word, 0][-1] == 'a':
        #print(df_emotion.iloc[word, 0])
        before_change = df_emotion.iloc[word, 0]       
        changing = list(before_change)
        changing[-1] = 'o'
        after_change = "".join(changing)
        #print(after_change)
        new_row = {'Spanish-es':after_change, 'emotion':df_emotion.iloc[word, 1]}
        df_emotion = df_emotion.append(new_row, ignore_index=True)
    
    
for word in range(len(df_emotion.index)):    
    if df_emotion.iloc[word, 0][-1] == 'o' or df_emotion.iloc[word, 0][-1] == 'a':
        word_plural = df_emotion.iloc[word, 0]+'s'
        #print(word_plural)
        new_row = {'Spanish-es':word_plural, 'emotion':df_emotion.iloc[word, 1]}
        df_emotion = df_emotion.append(new_row, ignore_index=True)
        
       
df_emotion = df_emotion.drop_duplicates(subset='Spanish-es', keep="first")
```
