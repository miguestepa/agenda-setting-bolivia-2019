```python
import pandas as pd

```

```python
name = 'eldiario_emotions'
```

```python
file = pd.read_csv(f'{name}.csv', encoding ='utf-8')
```

```python
df = pd.DataFrame(file)
```

```python
df
```

```python
indexNames = df[(df['n_trust'] == 0) & 
                (df['n_fear'] == 0) &
                (df['n_surprise'] == 0) &
                (df['n_joy'] == 0) &
                (df['n_sadness'] == 0) &
                (df['n_anger'] == 0) &
                (df['n_anticipation'] == 0) &
                (df['n_disgust'] == 0)
               ].index 

df.drop(indexNames , inplace=True)

df = df.drop(columns = ['Unnamed: 0'])

df = df.reset_index(drop=True)
```

```python
df
```

```python
df.to_csv(f'{name}_cleaned.csv', encoding ='utf-8')
```
